```python
def top_10(data, title):
    top_10_gross = df2[['Title',data]]
    TOP =  top_10_gross.sort_values(by=data ,ascending=False)
    TOP = TOP[0:15]
    ax  = TOP.plot(kind='barh', figsize=(10,7) , title= title)
    ax.set_yticklabels(list(TOP['Title']), minor=False)
    plt.ylabel('The name of the movie' ,fontsize=16)
    plt.xlabel(data ,fontsize=16)
    ax.yaxis.set_tick_params(labelsize='large')
```
